```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('/content/drive/MyDrive/Pandas/Car Price.xlsx')
df
```

```python
plt.xlabel("Mileage", color="red")
plt.ylabel("Price", color="red")
plt.scatter(df.Mileage,df.Price)
```

```python
lr= linear_model.LinearRegression()
lr.fit(df[['Mileage']],df.Price)
```

```python
lr.predict([[50000]])
```

```python
lr.score(df[['Mileage']],df.Price)
lr.coef_
```

```python
lr.intercept_
```

```python

predicted = 0.00133308 * 50000 + 1406.181763242865
print(predicted)
```

```python
plt.xlabel("Mileage", color="red")
plt.ylabel("Price", color="red")
plt.scatter(df.Mileage,df.Price)
plt.plot(df.Mileage, lr.predict(df[['Mileage']]),color='red')
```
